### Determining pareto distribution using log-log plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# DEFINING PARAMETERS OF PARETO DISTRIBUTION
alpha = 3
xm = 1

# CREATING ARRAY OF X VALUES
x = np.linspace(0.1,10,1000)

#CALCULATING Y-VALUES OF PARETO DISTRIBUTION
y = alpha * (xm**alpha) / (x**(alpha+1))

# CREATING PARETO DISTRIBUTION PLOT
plt.plot(x,y)
plt.show()
```

```python
# CREATING LOG-LOG PLOT TO DETERMINE THE PARETO DISTRIBUTION USING X AND Y VALUES
plt.plot(np.log(x),np.log(y))


plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Log-log plot for pareto distribution')

plt.show()
```
